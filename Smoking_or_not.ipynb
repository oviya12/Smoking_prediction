Importing the necessary libraries:

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### **DATA UNDERSTANDING:**

```python
df=pd.read_csv("smokin2_data.csv",header= None)
```

```python
df.head()
```

The top 5 rows are printed, which displays the 4 data rows and the header row as data row, as we have mentioned header as none during reading csv file

```python
df.describe()
```

```python
df.shape
```

Totally, our data set contains 500 rows and 26 columns

```python
print(df)
```

```python
new_df=df.drop(0)
```

Now, we have dropped the 0th row, that is our header row so that now our dataset contains only the numeric values

```python
new_df.head()
```

```python
new_df[25].value_counts()
```

In our dataset 25th column specify whether a person has smoking symptoms or not, Hence when we print the value count, it displays as 294 records with smoking symptoms and 205 records with no smoking symptoms

```python
new_df.groupby(25).mean()
```

```python
x=new_df.drop(columns=25,axis=0)
y=new_df[25]
print(x)
print(y)
```

x has been assigned with all the data required to find the result and the y has been assigned with the result

### **TRAINING AND TESTING DATA:**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

Now we split the data into test data and train data, and store it in corresponding varaiable, so that using the test data our model will be trained to produce result with some accuracy score, and the we can validate using
test data.

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

Since we need our model to predict the target outcome which is binary in nature (smoking or not smoking) we train our model with logistic regression



### **BUILDING MODEL:**

```python
model.fit(x_train,y_train)
```

Now we include the training data to our model to train the model to predict the outcome

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy: ",training_data_accuracy)
```

we can find the accuracy of our model prediction using accuracy score function, which is 74 % for our case

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy: ",testing_data_accuracy)
```

Accuracy for test prediction is 68%, which somewhat ok, as building a model that predicts with 100% accuracy is impractical

### **EVALUATING OUR MODEL:**

```python
input_data = (30,35,170,65,84.6,1,1,1,1,127,72,87,179,76,52,112,16,1,0.8,16,15,12,1,0,0,)
input_arr = np.asarray(input_data)
input_arr_reshaped = input_arr.reshape(1,-1)
prediction = model.predict(input_arr_reshaped)
if prediction[0]=='1':
  print("Has a Symptoms of smoking")
else:
  print("Has no symptoms of smoking")
```

Now a sample input is given to pour model which is first converted to array and then reshaped, then finally predicts the result
